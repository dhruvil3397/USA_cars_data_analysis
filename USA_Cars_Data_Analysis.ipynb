```python
data = []
for br in df["brand"].unique():
  data.append([br,len(df[df["brand"]==br])])

data = pd.DataFrame(data,columns=["brand","freq"])
data = data.sort_values(by="freq",ascending= False)
data.plot(x = "brand",y = "freq",kind = "bar",figsize=(30,10),title = "Maximum number of cars by brand ")
```

```python
plt.figure(figsize=(20,5))
plt.bar(data.brand,data.freq)
plt.xlabel("Brand")
plt.ylabel("Frequency")
```

**Q3. Car Industry Trend**

```python
df["year"].unique()
```

```python
data = []
for yr in df["year"].unique():
  data.append([yr,len(df[df["year"]==yr])])
df = pd.DataFrame(data,columns=["year","freq"])
data=df.sort_values(by="year",ascending=True)
plt.figure(figsize=(10,5))
plt.plot(data.year,data.freq)
plt.xlabel("Year")
plt.ylabel("Frequency")
plt.show()
```

**Q 5. Which car brand is the most Expensive**

```python
data= df[["brand","price"]]
data = pd.DataFrame(data,columns= ["brand","price"],)
data = data.sort_values(by="price",ascending = False).head(5)
data.plot(x = "brand",y = "price",kind = "bar",figsize=(15,8))
```

**Q 6. Which car brand is the most Expensive**

```python
data = df.groupby("brand").mean()["price"]
data = data.sort_values(ascending=False)
plt.figure(figsize=(10,5))
data.plot.bar()
```

**Q6. Which State is having most cars**

```python
df["state"].unique()
```

```python
data = []
for st in df["state"].unique():
  data.append([st,len(df[df["state"]==st])])
data = pd.DataFrame(data,columns = ["state","Frequency"])
data =data.sort_values(by="Frequency",ascending=False)
plt.figure(figsize=(20,8))
plt.bar(data.state,data.Frequency,)
plt.xlabel("State")
plt.ylabel("Frequency")
plt.show()
```
